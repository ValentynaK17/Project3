```python
# Dependencies
import requests
import pandas as pd
from pprint import pprint
```

```python
booksSmall_df = pd.read_csv('./Resources/BooksO.csv')
```

```python
#Creating dataframes from csv files to read the data
# books_df = pd.read_csv('./Resources2/Books.csv')
books_df = booksSmall_df[booksSmall_df['Year-Of-Publication'] != 0]
books_df=books_df[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
users_df = pd.read_csv('./Resources/UsersO.csv')
ratings_df = pd.read_csv('./Resources/RatingsO.csv')
geoloc_df = pd.read_csv('./Resources/GeoLocO.csv')
```

```python
books_df.head()
```

```python
books_df.info()
```

```python
books_df['Book-Title'] = books_df['Book-Title'].apply(lambda x: x.title())
```

```python
books_df=books_df.drop_duplicates(subset=['Book-Title'])
```

```python
books_df[books_df.duplicated(subset=['Book-Title'])]
```

```python
merged_books_df=books_df.merge(ratings_df,on='ISBN',how='left')
merged_books_df
```

```python
merged_books_df = merged_books_df[merged_books_df['Book-Author'].notna()]
merged_books_df
```

```python
merged_books_df['is_asciiB'] = merged_books_df['Book-Author'].apply(lambda x: all(ord(char) < 128 for char in x))
merged_books_df['is_asciiB']
```

```python
merged_books_df=merged_books_df[merged_books_df['is_asciiB']==True]
merged_books_df
```

```python
merged_books_df['is_ascii'] = merged_books_df['Book-Title'].apply(lambda x: all(ord(char) < 128 for char in x))
merged_books_df['is_ascii']
```

```python
merged_books_df=merged_books_df[merged_books_df['is_ascii']==True]
merged_books_df
```

```python
titles_list=merged_books_df['Book-Title']
list_t=list(titles_list.unique())
list_t
```

```python
books_df=books_df[books_df['Book-Title'].isin(list_t)]
books_df
```

```python
books_df[books_df['Book-Author'] == 'John Grisham']   
```

```python
ser=merged_books_df.groupby('Book-Author').size().sort_values(ascending=False)
ser_pd=pd.DataFrame(ser)
ser_pd
```

```python
merged_books_df.groupby('Book-Title').size().sort_values(ascending=False)
```
